```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# 1. URL
```

```python
url = 'http://corners.gmarket.co.kr/Bestsellers'
```

```python
# 2. request > response :HTML
```

```python
response = requests.get(url)
response
```

```python
response.text[:300]
```

```python
# 3. HTML(str) > BeautifulSoup Object > css-selector > DataFrame
```

```python
dom = BeautifulSoup(response.text,"html.parser")
type(dom)
```

```python
# 4. BeautifulSoup Object > css-selector > DataFrame
```

```python
# 200개 상품 엘리먼트 선택
elements = dom.select("#gBestWrap > div > div:nth-child(5) > div > ul > li")
len(elements)
```

```python
# 개발자 도구에서 element 선택 > 아이콘 클릭해서 선택 >오른쪽 마우스버튼, copy> copy selector
#gBestWrap > div > div:nth-child(5) > div > ul > li:nth-child(2)
```

```python
element = elements[0]
```

```python
data = {
    "title":element.select_one(".itemname").text,
    "o_price": element.select_one(".o-price").text,
    "s_price": element.select_one(".s-price > strong").text,
    "img": "http:" + element.select_one("img").get("data-original"),
}
data
```

```python
items = []
for element in elements:
    data = {
    "title":element.select_one(".itemname").text,
    "o_price": element.select_one(".o-price").text,
    "s_price": element.select_one(".s-price > strong").text,
    "img": "http:" + element.select_one("img").get("data-original"),
    }
    items.append(data)
items[:2]
```

```python
df = pd.DataFrame(items)
df.tail(2)
```
